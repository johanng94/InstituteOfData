# 7. [BONUS] Plot X, Y and Best Fit Line

**Write a function to calculate best fit.**

![alt text](https://i.imgur.com/B43vOV0.jpg)[1]

[1]: https://www.mathsisfun.com/data/least-squares-regression.html "Least Squares Regression"

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    
    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X)
    
    numerator = sum([x*y for x,y in zip(X,Y)]) - n*xbar * ybar
    denominator = sum([x**2 for x in X]) - n* xbar**2
    
    m = numerator/denominator
    b = ybar - m * xbar
    
    return m , b

# solution
m, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

# plot the data points and best fit line
plt.scatter(X,Y)
yfit = [b + m*x for x in X]
plt.plot(X,yfit)
plt.grid()
plt.show()
````

# 11. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

````python
import numpy as np
import scipy.stats

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
def mean_confidence_interval(data, confidence=0.95):
    # Calculate m (mean), se (std. error of mean), h (distance from mean to C.I.)
    m, se = np.mean(data),stats.sem(data)
    n = len(data)
    h = se*stats.t.ppf((1 + confidence)/2.,n-1)
    return m, m-h, m+h

print(mean_confidence_interval(data))
````
